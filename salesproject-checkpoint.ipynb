```python
total_sales_by_region = df.groupby('Region')['Sales'].sum().reset_index()
print("Total Sales by Region:")
print(total_sales_by_region)
```

```python
total_profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()
print("Total Profit by Category:")
print(total_profit_by_category)
```

```python
top_10_products = df.groupby('Product Name')['Sales'].sum().nlargest(10).reset_index()
print("Top 10 Products by Sales:")
print(top_10_products)
```

```python
top_10_profit = df.groupby('Product Name')['Profit'].sum().nlargest(10).reset_index()
print("Top 10 Products by profit:")
print(top_10_profit)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')

# Extract year and month from 'Order Date'
df['YearMonth'] = df['Order Date'].dt.to_period('M')

# Group by 'YearMonth' and calculate total sales
sales_trend = df.groupby('YearMonth')['Sales'].sum().reset_index()

print("Sales Trend Over Time:")
print(sales_trend)
```

```python
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
print("Profit Margin Analysis:")
print(df[['Product Name', 'Sales', 'Profit', 'Profit Margin']].head())
```

```python
# Set the style for seaborn
sns.set(style="whitegrid")

# Plot total sales by region
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=total_sales_by_region)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Convert 'YearMonth' back to string for plotting
sales_trend['YearMonth'] = sales_trend['YearMonth'].astype(str)

# Plot sales trend over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='YearMonth', y='Sales', data=sales_trend, marker='o')
plt.title('Sales Trend Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
